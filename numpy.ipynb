```python
import numpy as np
import matplotlib.pyplot as plt

x=np.array(["Apples","Banana","Mango","Grapes"])
y=np.array([4,15,20,2])

plt.barh(x,y,height=0.5, color="blue")
plt.show()
```

```python
x=np.array([0,1,2,3,4])
y=np.array([0,1,2,3,4])

plt.plot(x,y,marker="x")
plt.title("Title")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

```

```python
x_axis=np.array([0,3,5,7])
y_axis=np.array([0,5,8,16])

plt.scatter(x_axis,y_axis)
plt.show()
```

```python
x=np.array([210,90,50,10])
my_label=["BMW","Jeep","Toyota","Volvo"]
my_colors=["green","violet","black","yellow"]
my_explode=[0.3,0,0,0]

plt.pie(x,labels=my_label,colors=my_colors,explode=my_explode,shadow=True)
plt.legend(title="cars")
plt.show()
```

```python
from numpy import random

x=random.randint(100,size=(5))
y=random.randint(100,size=(5))
print(x)
print(y)

plt.plot(x,y)
plt.show
```

```python
x=random.randint(50,size=(5))
y=random.randint(100,size=(5))
print(x)
print(y)

plt.barh(x,y,height=0.5,color="blue")
plt.show
```

```python
from random import randint
blue=randint(0,100)
red=randint(0,100)
purple=randint(0,100)
white=randint(0,100)
x=np.array([blue,red,purple,white])
my_label=(["BMW","Toyota","Jeep","Volvo"])
myexplode=([0.3,0,0,0])
plt.pie(x,labels=my_label,explode=myexplode,shadow=True)
plt.legend(title="cars")

plt.show()
```
